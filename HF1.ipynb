```python
# Task 2.1
# Calculating the mean and the standard deviation of all the pixels
# Then normalizing the matrices for the pictures
imageList_corrected = list()
import numpy as np

for i in range(len(imageList)):
    # For the RGB channels, because we do the normalizing separately for each channel
    tmp = imageList[i][:, :, :]
    tmp2 = np.zeros((128, 128, 3))
    for j in range(3):
        mean = tmp[:, :, j].mean()
        std = tmp[:, :, j].std()
        step1 = (imageList[i][:, :, j] - mean) 
        step2 = step1 / step1.std()
        tmp2[:, :, j] = step2
    imageList_corrected.append(tmp2)
    
# Check whether the transformation was succesful 
for i in range(len(imageList_corrected)):
    for j in range(3):
        print("Mean of picture ", i, ", channel", j, ": " , imageList_corrected[i][:, :, j].mean())
        print("Std  of picture ", i, ", channel", j, ": " , imageList_corrected[i][:, :, j].std())




```

```python
# Task 3.1

import urllib.request
data = urllib.request.urlopen(("https://blog.keras.io/the-future-of-deep-learning.html"))
        
data_str = str(data.read())


# Completed with the help of:
# https://scipython.com/book/chapter-7-matplotlib/examples/letter-frequencies-in-moby-dick/
letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
# Initialize the dictionary of letter counts: {'A': 0, 'B': 0, ...}
lcount = dict([(l, 0) for l in letters])

# Count!
for l in data_str:
    try:
        lcount[l.upper()] += 1
    except KeyError:
        # Ignore characters that are not letters
        pass
    
# Plotting
x = letters
x_pos = [i for i, _ in enumerate(x)]
y = list(lcount.values())
# Beautify plot
plt.bar(x_pos, y, color='green')
plt.xticks(x_pos, x)
plt.xlabel("Letters (upper and lower case)")
plt.ylabel("Letter count")
```

```python
# Task 4
# based on: 
# https://www.kaggle.com/msripooja/steps-to-convert-audio-clip-to-spectrogram
import os
import matplotlib.pyplot as plt

#for loading and visualizing audio files
import librosa
import librosa.display

#to play audio
import IPython.display as ipd

audio_fpath = "/Users/szilard/Dropbox/MSc 1/Deep Learning/Jupyter/"
audio_clips = os.listdir(audio_fpath)

x, sr = librosa.load(audio_fpath+audio_clips[2], sr=44100)
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)

# Converting to spectogram
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()

# The same, but  on a logarithmic scale
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
```
